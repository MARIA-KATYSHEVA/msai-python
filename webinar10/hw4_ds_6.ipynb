```python
# Dari-Khanda Garmaeva solution
from datetime import datetime


class VendingMachine:
    time_now = datetime.now()
    current_time = int(time_now.strftime('%H'))
    items = ['coca_cola', 'water', 'orange_juice', 'gum', 'mints', 'chips', 'chocolate_bar', 'nuts', 'crackers']
    prices = [1, 1, 3, 0.25, 0.5, 1, 1, 1, 0.5]
    codes = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    def __init__(self, bank):
        self.bank = bank
        self.stock = [5, 0, 5, 5, 5, 5, 5, 5, 5]
        self.limit = 5
        self.secret_operator_code = 'Open Sesame'
        self.pincode = 1111

    def greeting(self):
        current_time = VendingMachine.current_time
        if current_time < 12:
            part_of_day = 'morning'
        elif 12 <= current_time < 17:
            part_of_day = 'afternoon'
        elif 17 <= current_time < 21:
            part_of_day = 'evening'
        else:
            part_of_day = 'night'

        print(f'Good {part_of_day}! Available items:')

    def available_items(self):
        for i in VendingMachine.codes:
            print(f'code {VendingMachine.codes[i]}:', f'{VendingMachine.items[i]}', f'${VendingMachine.prices[i]} ({self.stock[i]})')
        print('Please, pick an item and enter the code')

    def pick_item(self):
        item = input()
        return item

    def restock(self):
        first_answer = input('Are you an operator? True/False: ')
        if first_answer == 'True':
            second_answer = int(input('Please enter pincode: '))
            if second_answer == self.pincode:
                print('Please enter an item to restock:')
                item = int(input())
                if self.stock[item] == self.limit:
                    print('This item is not available for restock')
                else:
                    self.stock[item] += 1
                    self.available_items()
            else:
                print('Incorrect pincode, we are calling police...')
        else:
            print('Please start again')

    def get_cash(self, item):
        if item == self.secret_operator_code:
            self.restock()
        elif self.stock[int(item)] == 0:
            print('This item is out of stock, please choose another item')
        else:
            print(f'Please insert ${VendingMachine.prices[int(item)]}')
            cash = float(input())
            print(f'You inserted ${cash}')
            change = cash - VendingMachine.prices[int(item)]
            print(f'Your change is ${change}')
            self.bank += VendingMachine.prices[int(item)]
            self.stock[int(item)] -= 1

    def continue_purchasing(self):
        while True:
            self.available_items()
            self.get_cash(input())
            print('Would you like to continue? Yes/No')
            answer = input()
            if answer == 'No':
                break


if __name__ == '__main__':
    vm = VendingMachine(500)
    vm.available_items()
    vm.get_cash(vm.pick_item())
    vm.continue_purchasing()
```

```python
class ManageSyst:

    def __init__(self, owner, company_name, departments: list = None):
        self.owner = owner
        self.name = company_name
        self._data = {}
        self.set_departments(departments)

    def set_departments(self, departments: list = None):
        """
        :param departments: ["IT", "cleaning", ...other departments names]
        """
        if departments:
            for dep in departments:
                self._data[dep] = {"Boss": None, "employees": []}

    def set_bosses(self, departments: dict = None):
        """
        :param departments: example: {"IT": boss1, "cleaning": boss2}
        """
        for dep, boss in departments.items():
            self._data[dep]["Boss"] = boss

    def get_boss(self, dep_name):
        return self._data[dep_name]["Boss"]

    def get_workers(self, dep_name):
        return self._data[dep_name]["employees"]

    @staticmethod
    def create_worker(fullname: str, phone: str, email: str, position: str, salary: int,
                      history_of_growth: str, his_boss: str):
        all_worker_info = {"fullname": fullname, "phone": phone, "email": email, "position": position, "salary": salary,
                           "history_of_growth": history_of_growth, "his_boss": his_boss}
        return all_worker_info

    @staticmethod
    def print_worker(worker):
        return worker["fullname"]

    def add_employees(self, workers: dict):
        """
        :param workers: {"IT": [worker1, worker2, worker3, ...]}
        """
        for dep in workers:
            self._data[dep]["employees"].extend(workers[dep])

    def fire_worker(self, worker):
        """
        :param worker: dict of worker
        """
        his_department = ""
        for dep in self._data:
            if self._data[dep]["Boss"] == worker["his_boss"]:
                his_department = dep
        self._data[his_department]["employees"].remove(worker)

    def edit_workers(self, param, new_value, criterion):
        """
        :param param: это то, по какому параметру будет изменение, например salary или position
        :param new_value: новое значение param
        :param criterion: lambda функция, то есть условие по которому будут выбирать какую группу работников менять
        Например: lambda worker: worker['salary'] > 30000
        """
        for dep in self._data:
            for worker in self._data[dep]["employees"]:
                if criterion(worker):
                    worker[param] = new_value

    def select_workers(self, criterion):
        filtered_workers = []
        for dep in self._data:
            for worker in self._data[dep]["employees"]:
                if criterion(worker):
                    filtered_workers.append(worker)
        return filtered_workers
```

```python
# Vladislav Naumov solution
# %%writefile vladislav_naumov.py

from matplotlib import markers
import math
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def timer_for(length, function=None):
    """Takes lenght of list and math function
    return the time to change the lst list with for loop"""
    lst = list(range(length))
    # 2.1 YOUR_CODE_HERE
    start = datetime.datetime.now()  # datetime help here
    for i in range(length):
        lst[i] = function(lst[i])     # change lst with math function
    finish = datetime.datetime.now()
    #####################
    return finish - start


def timer_comp(length, function=None):
    """Takes lenght of list and math function
    return the time to change the lst list with comprehensions"""
    lst = list(range(length))
    # 2.2 YOUR_CODE_HERE
    start = datetime.datetime.now() # datetime help here
    lst = [function(lst[i]) for i in range(length)]
    finish = datetime.datetime.now()
    #####################
    return finish - start


def timer_np(length, ufunc=None):
    """Takes lenght of ndarray and np ufunc
    return the time to change ndarray with function"""
    arr = np.arange(length)
    # 2.3 YOUR_CODE_HERE
    start = datetime.datetime.now()  # datetime help here
    arr = ufunc(arr)         # change arr with ufunc
    finish = datetime.datetime.now()
    #####################
    return finish - start


def measurements(lengths):
    """Takes lengths of arrays
    return 2-d array (8, 4): stack lengths and measurements"""
    arr2d = np.zeros(shape=(8, 3))
    # 2.4 YOUR_CODE_HERE
    times_for = np.array([timer_for(l, math.log1p).microseconds for l in lengths])
    times_comp = np.array([timer_comp(l, math.log1p).microseconds for l in lengths])
    times_np = np.array([timer_np(l, np.log1p).microseconds for l in lengths])
    arr2d[:, 0] = times_for
    arr2d[:, 1] = times_comp
    arr2d[:, 2] = times_np
    arr2d = np.hstack((np.array(lengths).reshape(-1, 1), arr2d))
    #####################
    return arr2d


def create_df(arr2d):
    list_of_columns = ['length', 'for loop', 'compreh', 'numpy']
    # 3.1 YOUR_CODE_HERE
    df = pd.DataFrame(data=arr2d, columns=list_of_columns)
    df = df.set_index('length')
    #####################
    return df


def draw_and_save_plots(df, function=None, mail=None):
    plt.figure(figsize=(16, 10))
    # 4.1 YOUR_CODE_HERE
    markers = ['ro', 'g^', 'bs']
    colors = ['r', 'g', 'b']
    i = 0
    # for column in df.iloc[:, 1:].columns:
    for column in df.columns:
        plt.plot(df[column], markers[i], label=column)
        plt.plot(df[column], color=colors[i])
        i += 1
    plt.xlabel('length', fontsize=14)
    plt.ylabel('time, $\mu s$', fontsize=14)
    plt.grid()
    plt.legend(fontsize=12)
    ######################
    plt.title(f'{function} {mail}')
    return None

def main():
    # 1.1 YOUR_CODE_HERE
    name = 'name'
    surname = 'surname'
    mail = 'mail@phystech.edu'
    function = math.log1p  # 1.2
    ######################

    lengths = [10 ** i for i in range(8)]
    df = create_df(measurements(lengths))

    # 3.3 YOUR_CODE_HERE
    df.to_csv(f'time_perfomance_{name}_{surname}.csv')
    ######################

    draw_and_save_plots(df, function=function, mail=mail)
    plt.savefig(f'plot_{name}_{surname}.png')

    # 5.1 YOUR_CODE_HERE
    conclusion = """Numpy time performance per length is motonously 
    increasing like linear function. for-loop and list-comprehension 
    with math is working slower and time performance per length is not linear function."""
    ######################

if __name__ == '__main__':
    main()
```
